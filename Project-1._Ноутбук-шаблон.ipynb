1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
#ваш код здесь
# Cмотрим как выглядит запись 'Образование и ВУЗ' в ячейках одноименного столбца 
print(hh_df.iloc[5, 9])
print(hh_df.iloc[1121, 9])
print(hh_df.iloc[31057, 9])
# Зафиксируем количество строк для будущих проверок 
display(hh_df.shape[0])

print('======================================================================')

def seleсt_education(education):
    """Выбираем из строки с информацией об образовании и ВУЗе уровень образования  
    и определяем его в одну из 4-х категорий, согласно условиям задания.

    Args:
        education (str): Строка вида 'Высшее образование 2001  
        Тульский артиллерийский инженерный институт, Тула Радиоэлектронные системы'

    Returns:
        str: Высшее
        str: Неоконченное высшее
        str: Среднее специальное
        str: Среднее
    """
    # 1. Преобразуем полученный аргумент (str), в список строк по разделителю " " 
    education_split = education.split()
    
    # Выбираем из полученного списка 3 первые слова, при помощи условий 
    # определяем категорию
    education = education_split[0:3]
    if 'Высшее' in education:
        education_category = 'Высшее'
        
    elif 'Неоконченное' in education:
        education_category = 'Неоконченное высшее'
        
    elif 'Среднее' in education and 'специальное' in education:
        education_category = 'Среднее специальное'
        
    elif 'Среднее' in education:
        education_category = 'Среднее'
        
    return education_category
        
# Создаем новый признак 'Образование', преобразуя признак 'Образование и ВУЗ' 
# при помощи функции "seleсt_education".    
hh_df['Образование'] = hh_df['Образование и ВУЗ'].apply(seleсt_education)  

# Произведем проверку
display(hh_df['Образование'].value_counts())
display(hh_df['Образование'].shape[0]) 

print('======================================================================')
# Ответ на Задание 3.1 из юнита.
mask1 = hh_df['Образование'] == 'Среднее'
count_mean = hh_df[mask1].shape[0]
display(f'Средний уровень образования имеют: {count_mean} соискателей')

print('======================================================================')
# Удаляем признак 'Образование и ВУЗ' и проверяем факт удаления
hh_df = hh_df.drop('Образование и ВУЗ', axis=1)
hh_df.info()
```

```python
#ваш код здесь
# Cмотрим как выглядит запись 'опыт работы' в ячейках одноименного столбца 
#print(hh_df.loc[5, 'Опыт работы'])
#print(hh_df.loc[1121, 'Опыт работы'])
#print(hh_df.loc[31057, 'Опыт работы'])

# Перед началом работы над столбцом "Опыт работы" зафиксируем общее количество
# строк (пригодится для проверки).
display(hh_df['Опыт работы'].shape[0])

print('======================================================================')
# Выясняем, есть ли пропущенные значения в столбце "Опыт работы" и сколько их
display(hh_df['Опыт работы'].isnull().sum())

# Проверяем во всех ли заполненных ячейках, строка начинается со слов "Опыт работы"
hh_df['Опыт'] = hh_df['Опыт работы'].apply(
    lambda x: x.split()[:2] if type(x) is str else np.nan
)
display(hh_df['Опыт'].value_counts())
# Небольшой анализ показал, что все заполненные ячейки начинаются со слов
# "Опыт работы", ячйки, заполненные неправильно имеют два варинта заполнения 
# это: "Не указано" и пустая ячейка. Сумма этих трех вариантов равна количеству 
# строк в столбце "Опыт работы". 
# Теперь можно приступать к преобразованию столбца "Опыт работы"

print('======================================================================')

def get_experience(args):
    """Преобразуем строку со стажем работы в общее количество месяцев.

    Args:
        args (str): Строка вида "Опыт работы 5 лет 2 месяца".

    Returns:
        int: Общее количество месяцев стажа.
        np.nan: Если входные данные не являются строкой или стаж не указан.
    """
    # 1. Проверка типа и пустых значений
    if type(args) is not str: 
       return np.nan
    if args == 'Не указано':
        return np.nan
    
    # 2. Определение ключевых слов
    key_m = ['месяц', 'месяца', 'месяцев']
    key_y = ['год', 'года', 'лет']
    
    # 3. По индексам определяем значение слова, производим проверку 
    # и осуществляем необходимые вычисления.
    args_split = args.split()   
    if args_split[5].strip() in key_m:
        exp_months = int(args_split[2])*12 + int(args_split[4])
        
    elif args_split[3].strip() in key_y and args_split[5] not in key_m:
        exp_months = int(args_split[2])*12 
          
    elif args_split[3].strip() in key_m:
        exp_months = int(args_split[2])  
          
    return exp_months

# Создаем новый признак 'Опыт работы (месяц)' преобразуя признак 'Опыт работы', 
# с помощью функции "get_experience".
hh_df['Опыт работы (месяц)'] = hh_df['Опыт работы'].apply(get_experience)

# Проверяем получившийся результат
display(hh_df['Опыт работы (месяц)'])

print('======================================================================')
# Ответ на Задание 3.3
value_median = hh_df['Опыт работы (месяц)'].median()
display(f'Медианный опыт работы (в месяцах) в нашей таблице = {value_median}')

print('======================================================================')
# Удаляем не нужные признаки
hh_df = hh_df.drop(['Опыт работы', 'Опыт'], axis=1)
hh_df.info()  # Проверяем удаление

```

```python
#ваш код здесь
# Cмотрим как выглядит запись в ячейках признака 'Город, переезд, командировки'
print(hh_df.loc[5, 'Город, переезд, командировки'])
print(hh_df.loc[1121, 'Город, переезд, командировки'])
print(hh_df.loc[1722, 'Город, переезд, командировки'])

# Зафиксируем количество строк до преобразований (для проверки).
display(hh_df['Город, переезд, командировки'].shape[0])

# Поверяем наличие пропусков в столбце.
display(hh_df['Город, переезд, командировки'].isnull().sum())

print('======================================================================')

def division_by_cities(attribute):
    """Выбираем из строки с мнформацией о городе и других параметрах 
    - только город и определяем его в одну из 4-х категорий в cоответсвии 
    с условием задания.

    Args:
        attribute (str): Строка вида 'Москва ,  м. Полежаевская , 
        готов к переезду , готов к командировкам'

    Returns:
        str: 'Москва'
        str: 'Санкт-Петербург'
        str: 'город-миллионник'
        str: 'другие'
    """
    # 1. Преобразуем полученный аргумент (str), в список строк по разделителю ","
    attribute_lst = attribute.split(',')
    
    # 2. Создаем ключ для определения города-миллионника
    million_cities = [
                'Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань',
                'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа',
                'Красноярск', 'Пермь', 'Воронеж', 'Волгоград'
    ]
    # 3. По индексам, с помощью условий определяем категорию
    if attribute_lst[0].strip() == 'Москва':
        category_city = 'Москва'
        
    elif attribute_lst[0].strip() == 'Санкт-Петербург':
        category_city = 'Санкт-Петербург'
        
    elif attribute_lst[0].strip() in million_cities:
        category_city = 'город-миллионник'
        
    else:
        category_city = 'другие'
        
    return category_city

# Cоздаем новый признак 'Город' преобразуя признак 'Город, переезд, командировки', 
# с помощью функции "division_by_cities".
hh_df['Город'] = hh_df['Город, переезд, командировки'].apply(division_by_cities)  

# Осуществляем проверку.
display(hh_df['Город'].value_counts()) 

print('----------------------------------------------------------------------')  

def ready_to_relocate(status):
    """Выбираем из строки с информацией о городе,  готовностью к переезду
    и готовностью к командировкам - только готовность к переезду. 
    Возвращаем значение True или False, по принципу готов - не готов.

    Args:
        status (str): Строка вида 'Старый Оскол , не готов к переезду , 
        не готов к командировкам'

    Returns:
        bool: True, если условие выполнено, и False в противном случае.
    """
    # 1. Преобразуем полученный аргумент (str), в список строк по разделителю ","
    status_lst = status.split(',')    
    # 2. Определяем ключевые слова
    status_true = [
        'готов к переезду', 'готова к переезду', 
        'хочу переехать', 'хочу переезжать'
    ]
    status_false =  [
        'не готов к переезду', 'не готова к переезду', 
        'не хочу переехать', 'не хочу переезжать'
    ]
    
    # 3. С помощью цикла перебираем строки и сравниваем их с ключевыми словами, 
    # для понимания готов/не готов кандидат к переезду.  
    for elem in status_lst:
        if elem.strip() in  status_true:
            return  True
        
        elif elem.strip() in status_false:
            return False
        
        elif elem.split('(')[0].strip() in status_true:
            return True
        
        elif elem.split('(')[0].strip() in status_false:
            return False

# Cоздаем новый признак 'Готовность к переезду', преобразуя признак 
# 'Город, переезд, командировки', с помощью функции "ready_to_relocate".        
hh_df['Готовность к переезду'] = hh_df['Город, переезд, командировки'].apply(
    ready_to_relocate
)
# Осуществляем проверку.
display(hh_df['Готовность к переезду'].value_counts())

print('----------------------------------------------------------------------')

def ready_for_business_trips(status):
    """Выбираем из строки с информацией о городе,  готовностью к переезду
    и готовностью к командировкам - только готовность к командировкам. 
    Возвращаем значение True или False, по принципу готов - не готов.

    Args:
        status (str): Строка вида 'Москва ,  м. Полежаевская , готов к переезду , 
        готов к командировкам'
    Returns:
        bool: True, если условие выполнено, и False в противном случае.
    """
    # 1. Преобразуем полученный аргумент (str), в список строк по разделителю ","
    status_lst = status.split(',')
    # 2. Определяем ключевые слова
    status_true = [
        'готов к командировкам', 'готов к редким командировкам', 
        'готова к командировкам','готова к редким командировкам', 'хочу в командировку'
    ] 
    # 3. С помощью цикла перебираем строки и сравниваем их с ключевыми словами, 
    # для понимания готов/не готов кандидат к командировкам.   
    for elem in status_lst:
        if elem.strip() in  status_true:
            return  True
      
    return False

# Cоздаем новый признак 'Готовность к командировкам', преобразуя признак 
# 'Город, переезд, командировки', с помощью функции "ready_for_business_trips".
hh_df['Готовность к командировкам'] = hh_df['Город, переезд, командировки'].apply(
    ready_for_business_trips
)
# Осуществляем проверку.
display(hh_df['Готовность к командировкам'].value_counts())
hh_df.info()  # на создание новых признаков в df

print('======================================================================')
# Ответы на Задание 3.4 из юнита:
# Вопрос №1
hh_piter = hh_df['Город'].value_counts()
piter = round(hh_piter['Санкт-Петербург']*100 / hh_piter.sum())
print(f'В Санкт-Петербурге живут: {piter} % соискателей от всех представленых резюме')
# Вопрос №2
mask1 = hh_df['Готовность к командировкам'] == True
mask2 = hh_df['Готовность к переезду'] == True
ready_both_percent = round(hh_df[mask1 & mask2].shape[0]*100 / hh_df.shape[0])
print(
    f'Количество претендентов, готовых и к переезду, и к командировкам: '
    f'{ready_both_percent}%'
)
print('======================================================================')
# Удаляем призак 'Город, переезд, командировки'
hh_df = hh_df.drop(['Город, переезд, командировки'], axis=1) 
hh_df.info()  # Проверяем удаление 
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь
# Cмотрим как выглядит запись в ячейках признака 'Занятось'
print(hh_df.loc[5, 'Занятость'])
print(hh_df.loc[1121, 'Занятость'])
print(hh_df.loc[1722, 'Занятость'])

# Поверяем наличие пропусков в столбце.
display(hh_df['Занятость'].isnull().sum())

print('======================================================================')
# Определяем список с названиями новых признаков-мигалок по методу 
# One Hot Encoding, для нашего DF
get_names_col = []
for elem in hh_df['Занятость']:
    elem = elem.split(',')  # Преобразуеv str, в список строк по разделителю ","
     
    for cell in elem:  # ОИтерируем полученный список
        cell = cell.strip()  # Очистка значения от пробелов и т.п.
        
        if cell not in get_names_col:  # Проверяем на уникальность
            get_names_col.append(cell)  # Уникальные значения заносим в список.
       
print(get_names_col)

print('======================================================================')
# С помощью цикла и Лямбда-функции создаем новые признаки-мигалки в hh_df
for name in get_names_col:
    # Лямбда-функция проверяет типы занятости и заполняет ячейки 
    # соответствующего признака значениями True/False 
    # по принципу подходит/не подходит.
    hh_df[name] = hh_df['Занятость'].apply(
        lambda x: name in [item.strip() for item in x.split(',')]
    )
# Проверяем полученный результат
display(hh_df.head())
print(hh_df.info())

print('======================================================================')

# Cмотрим как выглядит запись в ячейках признака 'График'
print(hh_df.loc[5, 'График'])
print(hh_df.loc[1121, 'График'])
print(hh_df.loc[1722, 'График'])

# Поверяем наличие пропусков в столбце.
display(hh_df['График'].isnull().sum())

print('======================================================================')
# Определяем список с названиями новых признаков-мигалок по методу 
# One Hot Encoding, для нашего DF
get_names_col2 = []
for elem2 in hh_df['График']:
    elem2 = elem2.split(',') 
      
    for cell2 in elem2:
        cell2 = cell2.strip()
        
        if cell2 not in get_names_col2:
            get_names_col2.append(cell2 )
       
print(get_names_col2)

print('======================================================================')
# С помощью цикла и Лямбда-функции создаем новые признаки-мигалки в hh_df
for name2 in get_names_col2:
    # Лямбда-функция проверяет какой график подходит соискателю 
    # и заполняет ячейки соответствующего признака значениями True/False 
    # по принципу подходит/не подходит.
    hh_df[name2] = hh_df['График'].apply(
        lambda x: name2 in [item2.strip() for item2 in x.split(',')]
    )
# Проверяем полученный результат
display(hh_df.head())
print(hh_df.info())

print('======================================================================')
# Ответ на Задание 3.5
# Вопрос №1
mask2 = hh_df['проектная работа'] == True
mask3 = hh_df['волонтерство'] == True
both_sign = hh_df[mask2 & mask3].shape[0]
display(f'Проектную работу и волонтерство ищут: {both_sign} человек')
# Вопрос №2
mask3 = hh_df['вахтовый метод'] == True
mask4 = hh_df['гибкий график'] == True
both_sign2 = hh_df[mask3 & mask4].shape[0]
display(f'Вахтовым методом и с гибким графиком хотят работать: {both_sign2} человек')

print('======================================================================')
# Удаляем не нужные строки  и проверяем итоговый результат
hh_df = hh_df.drop(['График', 'Занятость'], axis=1)

hh_df.info()
display(hh_df.head())
print(hh_df.shape)
```

* close — цена закрытия (последний зафиксированный курс валюты на указанный день).

```python
#ваш код здесь
ExchangeRates = pd.read_csv('data/ExchangeRates.csv', sep = ',')
exchange_df = ExchangeRates.copy()
#display(exchange_df.head())
#exchange_df.info()
# Удаляем не нужные признаки 
exchange_df = exchange_df.drop(['time', 'vol', 'per'], axis=1)

# Cмотрим как выглядит запись в ячейках признака 'ЗП'
print(hh_df.loc[5, 'ЗП'])
print(hh_df.loc[1121, 'ЗП'])
print(hh_df.loc[1722, 'ЗП'])

# Поверяем наличие пропусков в столбце.
display(hh_df['ЗП'].isnull().sum())

print('======================================================================')

# Переводим признак 'Обновление резюме' из таблицы "hh_df" в формат datetime 
# и оставим только дату.
hh_df['Обновление резюме'] = pd.to_datetime(hh_df['Обновление резюме'], dayfirst=True)
hh_df['Обновление резюме'] = hh_df['Обновление резюме'].dt.normalize()

# Переводим признак 'date' из таблицы "exchange_df" в формат datetime
exchange_df['date'] = pd.to_datetime(exchange_df['date'], dayfirst=True)

# Проверим результат преобразований с датами в обеих таблицах
display(hh_df.head())
display(exchange_df.head())

print('======================================================================')

# Cоздадим новые признаки: 'ЗП_размер' и 'ЗП_валюта' преобразуя признак 'ЗП', 
# с помощью Лямбда-функций. Признак 'ЗП_размер' получит значения 
# int: Числовое значение з/п, а признак 'ЗП_валюта' получит значения 
# str: Обозначение валюты.
hh_df['ЗП_размер'] = hh_df['ЗП'].apply(lambda x: int(x.split()[0].strip()))
hh_df['ЗП_валюта'] = hh_df['ЗП'].apply(lambda x: x.split()[1].strip())

# Сравним две таблицы на предмет совпадения обозначения валют
display(hh_df['ЗП_валюта'].value_counts())

print('----------------------------------------------------------------------')
display(exchange_df['currency'].value_counts())
# Вывод: названия 3-х валют в таблицах не совпадают

print('----------------------------------------------------------------------')
# Перед объединением двух таблиц необходимо преобразовать значения в признаке  
# hh_df['ЗП_валюта'], к принятому стандарту. Это обеспечит совпадения значений 
# в двух представленных здесь таблицах.
def change_valuta(val):
    """Приведем обозначения валют к общепринятым стандартам.

    Args:
        val (str): Строка вида 'бел.руб.' или 'грн.' или 'сум'

    Returns:
        str: 'BYN'
        str: 'UAH'
        str: 'UZS'
    """
    if val == 'бел.руб.':
        return 'BYN'
    elif val == 'грн.':
        return 'UAH'
    elif val == 'сум':
        return 'UZS'
    else:
        return val

hh_df['ЗП_валюта'] = hh_df['ЗП_валюта'].apply(change_valuta)

# Проверим получившийся результат
display(hh_df['ЗП_валюта'].value_counts())

print('======================================================================')
# Перед объединеним таблиц, признаки, по которым будет производиться слияние, 
# рекомендуется привести к единому наименованию».
# Менять наименования признаков лучше в таблице "exchange_df" в связи с тем, 
# что преобразования в ней были минимальные в отличие от таблицы "hh_df".    
exchange_df = exchange_df.rename(columns={'date': 'Обновление резюме'})
exchange_df = exchange_df.rename(columns={'currency': 'ЗП_валюта'})

# Убедимся, что переименование прошло успешно
display(exchange_df.head(2))

print('======================================================================')
# Объединим две таблицы в одну под названием "merge_df". Объединение будем 
# производить по двум ключевым столбцам 'ЗП_валюта' и 'Обновление резюме'
merge_df = hh_df.merge(exchange_df, on = ['ЗП_валюта', 'Обновление резюме'], how='left')

# Посмотрим на результат объединения
display(merge_df.head())
display(merge_df.shape)
merge_df.info()

print('======================================================================')
# Заполним пустые ячейки признаков 'close' и 'proportion' 
# в новой таблице "merge_df"
mask5 = merge_df['ЗП_валюта'] == 'руб.'
merge_df.loc[mask5, 'close'] = 1
merge_df.loc[mask5, 'proportion'] = 1

# Проверим полученный результат
display(merge_df.head())

print('======================================================================')
# При необходимости можно активировать проверку на пустые ячейки
#merge_df['ЗП_валюта'].isnull().sum()
#merge_df['ЗП_размер'].isnull().sum()
#merge_df['close'].isnull().sum()

# Cоздаем новый признак 'ЗП (руб)'. Значения в данном признаке будут приведены 
# к единой валюте — российскому рублю.
merge_df['ЗП (руб)'] = merge_df['ЗП_размер'] * merge_df['close'] / merge_df['proportion']

# Производим проверку заполнения ячеек признака 'ЗП (руб)'.
display(merge_df['ЗП (руб)'])

print('======================================================================')
# Ответ на Задание 3.6
median_sel = round(merge_df['ЗП (руб)'].median() / 1000)
print(
    f'Желаемая медианная заработная плата соискателей в нашей таблице ' 
    f'составляет: {median_sel} тыс. руб.'
)
print('======================================================================')
# Удаляем не актульные признаки
merge_df = merge_df.drop('ЗП', axis=1)
merge_df.info()   # проверяем удаление
```

```python
def outliers_z_score_mod(data, feature, log_scale=False, left=3, right=3):

    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

outliers, cleaned = outliers_z_score_mod(df_plot, 'log_ЗП', left=3, right=4)
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')
```

Найдем выбросы с помощью метода z-отклонения

```python
def outliers_z_score_mod(data, feature, log_scale=False, left=3, right=3):

    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

outliers, cleaned = outliers_z_score_mod(hh_clean, 'Возраст', log_scale=True, left=3, right=4)

# Ответ на Задание 5.6
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')
```
